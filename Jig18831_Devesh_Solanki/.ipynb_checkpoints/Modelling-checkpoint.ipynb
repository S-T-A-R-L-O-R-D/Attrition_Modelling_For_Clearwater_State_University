# Building Bagged Classifier To know more about Variable affect on retention
 

```python
data=pd.get_dummies(data)
y=data['RETURNED_2ND_YR']
x=data.drop(['RETURNED_2ND_YR','STUDENT.IDENTIFIER'],axis=1)
from sklearn.model_selection import train_test_split as tts
train_x,test_x,train_y,test_y=tts(x,y,test_size=0.2,random_state=200)
```

```python
from sklearn.ensemble import BaggingClassifier as BGC
from sklearn.tree import DecisionTreeClassifier as DTC
clf=BGC(oob_score=True,n_jobs=-1,n_estimators=40,base_estimator=DTC())
clf.fit(train_x,train_y)
clf.oob_score_
```

```python
print(metrics.roc_auc_score(test_y,clf.predict_proba(test_x)[:,1]))
fpr,tpr,threshold=metrics.roc_curve(test_y,clf.predict_proba(test_x)[:,1])
plt.plot(fpr,tpr,"-")
```

```python
for w in range(30,210,20):
    clfTest=BGC(n_estimators=w,oob_score=True,n_jobs=-1,base_estimator=DTC())
    clfTest.fit(train_x,train_y)
    print(f"oob score for {w} estimators is {clfTest.oob_score_}\n")
```

```python
clf=BGC(oob_score=True,n_jobs=-1,n_estimators=150,base_estimator=DTC())
clf.fit(train_x,train_y)
acc_bag=round(100*clf.oob_score_,2)
print(acc_bag)
print(clf.oob_score_)
```

```python
auc_bag=metrics.roc_auc_score(test_y,clf.predict_proba(test_x)[:,1])
```

```python
print(metrics.roc_auc_score(test_y,clf.predict_proba(test_x)[:,1]))
fpr,tpr,threshold=metrics.roc_curve(test_y,clf.predict_proba(test_x)[:,1])
plt.plot(fpr,tpr,"-")
```

```python
imp=[]
for i in clf.estimators_:
    imp.append(i.feature_importances_)
imp=np.mean(imp,axis=0)
```

```python
imp=imp[0:20]
imp
```

```python
important_features=pd.Series(imp,index=x.columns[0:20].tolist())
important_features.sort_values(ascending=False).plot(kind='bar')
```

## stacking ensembles

```python
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier,AdaBoostClassifier as ABC ,BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
models=[ 
            ######## First level ########
            [RandomForestClassifier (n_estimators=150, criterion="entropy", max_depth=5, max_features=0.5, random_state=1),
             ExtraTreesClassifier (n_estimators=150, criterion="entropy", max_depth=5, max_features=0.5, random_state=1),
             GradientBoostingClassifier(n_estimators=150, learning_rate=0.6, max_depth=5, max_features=0.5, random_state=1),
             LogisticRegression(random_state=1), KNeighborsClassifier(n_neighbors = 2),
             BaggingClassifier(n_estimators=150,oob_score=True,n_jobs=-1,base_estimator=DTC())
             ],
            ######## Second level ########
            [
                GaussianNB()
            ]
        ]
```

```python
from pystacknet.pystacknet import StackNetClassifier

model=StackNetClassifier(models, metric="accuracy", folds=4,
    restacking=True,use_retraining=False, use_proba=True, 
    random_state=100,n_jobs=1, verbose=1)

model.fit(train_x,train_y)
```

```python
preds=model.predict_proba(test_x)
```

```python
import numpy as np
Y_preds=(np.argmax(preds,axis=1))
```

```python
round(((Y_preds==test_y).sum()/test_y.shape[0])*100,2)
```

```python
metrics.roc_auc_score(test_y,model.predict_proba(test_x)[:,1])
```
