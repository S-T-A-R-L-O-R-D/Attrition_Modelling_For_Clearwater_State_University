## drop variables that have quit high number of missing values (>25%) Or too many categories

```r
data=select(data,-c('CORE_COURSE_NAME_1_F','CORE_COURSE_NAME_2_F','CORE_COURSE_NAME_3_F','CORE_COURSE_NAME_4_F',
                'CORE_COURSE_NAME_5_F','CORE_COURSE_NAME_6_F','CORE_COURSE_GRADE_6_F','CORE_COURSE_GRADE_5_F',
                'CORE_COURSE_GRADE_4_F','CORE_COURSE_NAME_4_S','CORE_COURSE_NAME_5_S','CORE_COURSE_NAME_6_S',
                'CORE_COURSE_GRADE_6_S','CORE_COURSE_GRADE_5_S','CORE_COURSE_GRADE_4_S','CORE_COURSE_NAME_3_S',
                'CORE_COURSE_GRADE_3_S','CORE_COURSE_NAME_2_S','CORE_COURSE_NAME_1_S','STDNT_TEST_ENTRANCE1',
                'STDNT_TEST_ENTRANCE2','HIGH_SCHL_NAME','FATHER_HI_EDU_CD','MOTHER_HI_EDU_CD','DEGREE_GROUP_CD','STDNT_MAJOR',
                   'STDNT_MINOR'))
```

## Missing value imputation on the basis of distribution of data

```r
data$MOTHER_HI_EDU_DESC[which(data$MOTHER_HI_EDU_DESC=="No information")]<-0
data$FATHER_HI_EDU_DESC[which((data$FATHER_HI_EDU_DESC=="No information"))]<-0

data$STDNT_TEST_ENTRANCE_COMB[is.na(data$STDNT_TEST_ENTRANCE_COMB)]<-(950+990)/2
data$DISTANCE_FROM_HOME[is.na(data$DISTANCE_FROM_HOME)]<-mean(data$DISTANCE_FROM_HOME,na.rm=TRUE)
data$HIGH_SCHL_GPA[is.na(data$HIGH_SCHL_GPA)]<-(2.94+3.07)/2
data$SECOND_TERM_ATTEMPT_HRS[is.na(data$SECOND_TERM_ATTEMPT_HRS)]<-mean(data$SECOND_TERM_ATTEMPT_HRS,na.rm=TRUE)
data$SECOND_TERM_EARNED_HRS[is.na(data$SECOND_TERM_EARNED_HRS)]<-mean(data$SECOND_TERM_EARNED_HRS,na.rm=TRUE)
data$FIRST_TERM_ATTEMPT_HRS[is.na(data$FIRST_TERM_ATTEMPT_HRS)]<-mean(data$FIRST_TERM_ATTEMPT_HRS,na.rm=TRUE)
data$FIRST_TERM_EARNED_HRS[is.na(data$FIRST_TERM_EARNED_HRS)]<-mean(data$FIRST_TERM_EARNED_HRS,na.rm=TRUE)
data$STDNT_AGE[is.na(data$STDNT_AGE)]<-18
data$GROSS_FIN_NEED[is.na(data$GROSS_FIN_NEED)]<-1180020
data$COST_OF_ATTEND[is.na(data$COST_OF_ATTEND)]<-1264860
data$UNMET_NEED[is.na(data$UNMET_NEED)]<-462000
```

## Clubbing together similar categories into one

```r
data$CORE_COURSE_GRADE_1_F[which(data$CORE_COURSE_GRADE_1_F=="missing"|data$CORE_COURSE_GRADE_1_F=="NOT REP"|
                          data$CORE_COURSE_GRADE_1_F=="B"|data$CORE_COURSE_GRADE_1_F=="A")]<-"High"
data$CORE_COURSE_GRADE_1_F[is.na(data$CORE_COURSE_GRADE_1_F)]<-"Mid"
data$CORE_COURSE_GRADE_1_F[which(data$CORE_COURSE_GRADE_1_F=="C"|data$CORE_COURSE_GRADE_1_F=="D")]<-"Mid"
data$CORE_COURSE_GRADE_1_F[which(data$CORE_COURSE_GRADE_1_F=="F"|data$CORE_COURSE_GRADE_1_F=="INCOMPL")]<-"Low"

data$CORE_COURSE_GRADE_2_F[which(data$CORE_COURSE_GRADE_2_F=="missing"|data$CORE_COURSE_GRADE_2_F=="NOT REP"|
                          data$CORE_COURSE_GRADE_2_F=="B"|data$CORE_COURSE_GRADE_2_F=="A")]<-"High"
data$CORE_COURSE_GRADE_2_F[is.na(data$CORE_COURSE_GRADE_2_F)]<-"Mid"
data$CORE_COURSE_GRADE_2_F[which(data$CORE_COURSE_GRADE_2_F=="C"|data$CORE_COURSE_GRADE_2_F=="D")]<-"Mid"
data$CORE_COURSE_GRADE_2_F[which(data$CORE_COURSE_GRADE_2_F=="F"|data$CORE_COURSE_GRADE_2_F=="INCOMPL")]<-"Low"


data$CORE_COURSE_GRADE_3_F[is.na(data$CORE_COURSE_GRADE_3_F)]<-"Mid"
data$CORE_COURSE_GRADE_3_F[which(data$CORE_COURSE_GRADE_3_F=="A"|data$CORE_COURSE_GRADE_3_F=="B"|
                          data$CORE_COURSE_GRADE_3_F=="missing")]<-"High"
data$CORE_COURSE_GRADE_3_F[which(data$CORE_COURSE_GRADE_3_F=="C"|data$CORE_COURSE_GRADE_3_F=="D")]<-"Mid"
data$CORE_COURSE_GRADE_3_F[which(data$CORE_COURSE_GRADE_3_F=="F"|data$CORE_COURSE_GRADE_3_F=="INCOMPL")]<-"Low"

data$CORE_COURSE_GRADE_2_S[is.na(data$CORE_COURSE_GRADE_2_S)]<-"Mid"
data$CORE_COURSE_GRADE_2_S[which(data$CORE_COURSE_GRADE_2_S=="A"|data$CORE_COURSE_GRADE_2_S=="B"|
                          data$CORE_COURSE_GRADE_2_S=="missing")]<-"High"
data$CORE_COURSE_GRADE_2_S[which(data$CORE_COURSE_GRADE_2_S=="C"|data$CORE_COURSE_GRADE_2_S=="D")]<-"Mid"
data$CORE_COURSE_GRADE_2_S[which(data$CORE_COURSE_GRADE_2_S=="F"|data$CORE_COURSE_GRADE_2_S=="INCOMPL")]<-"Low"


data$CORE_COURSE_GRADE_1_S[is.na(data$CORE_COURSE_GRADE_1_S)]<-"Mid"
data$CORE_COURSE_GRADE_1_S[which(data$CORE_COURSE_GRADE_1_S=="A"|data$CORE_COURSE_GRADE_1_S=="B"|
                                 data$CORE_COURSE_GRADE_1_S=="NOT REP")]<-"High"
data$CORE_COURSE_GRADE_1_S[which(data$CORE_COURSE_GRADE_1_S=="C"|data$CORE_COURSE_GRADE_1_S=="D")]<-"Mid"
data$CORE_COURSE_GRADE_1_S[which(data$CORE_COURSE_GRADE_1_S=="F"|data$CORE_COURSE_GRADE_1_S=="INCOMPL"
                                |data$CORE_COURSE_GRADE_1_S=="missing"|data$CORE_COURSE_GRADE_1_S=="Unknown")]<-"Low"

data$STDNT_BACKGROUND[which(data$STDNT_BACKGROUND=="BGD 1"|data$STDNT_BACKGROUND=="BGD 2")]<-"BGD HIGH"
data$STDNT_BACKGROUND[which(data$STDNT_BACKGROUND=="BGD 3"|data$STDNT_BACKGROUND=="BGD 4"
                      |data$STDNT_BACKGROUND=="BGD 6"|data$STDNT_BACKGROUND=="BGD 8")]<-"BGD MID"
data$STDNT_BACKGROUND[which(data$STDNT_BACKGROUND=="BGD 5"|data$STDNT_BACKGROUND=="BGD 7")]<-"BGD LOW"
                       
```

## Converting continuous data into categorical on the basis of generated profiles

```r
data$COST_OF_ATTEND<-cut(data$COST_OF_ATTEND,breaks=c(0,150540,1264860,2124900),right=FALSE,include.lowest=T,labels=c('low','mid','high'))

#data$DISTANCE_FROM_HOME<-cut(data$DISTANCE_FROM_HOME,breaks=c(0,59,72,6000),right=FALSE,include.lowest=T,labels=c('low','mid','high'))

data$EST_FAM_CONTRIBUTION<-cut(data$EST_FAM_CONTRIBUTION,breaks=c(0,167940,1179600,5999940),right=FALSE,include.lowest=T,labels=c('low','mid','high'))

data$GROSS_FIN_NEED<-cut(data$GROSS_FIN_NEED,breaks=c(0,301440,1180020,2124900),right=FALSE,include.lowest=T,labels=c('low','mid','high'))

data$HIGH_SCHL_GPA<-cut(data$HIGH_SCHL_GPA,breaks=c(0,2.94,3.2,4.1),right=FALSE,include.lowest=T,labels=c('low','mid','high'))
data$STDNT_AGE<-cut(data$STDNT_AGE,breaks=c(16,18,26),right=FALSE,include.lowest=T,labels=c('low','mid'))

#data$STDNT_TEST_ENTRANCE_COMB<-cut(data$STDNT_TEST_ENTRANCE_COMB,breaks=c(530,910,1110,1590),right=FALSE,include.lowest=T,labels=c('low','mid','high'))

data$UNMET_NEED<-cut(data$UNMET_NEED,breaks=c(-1212072,20160,462000,1632660),right=FALSE,include.lowest=T,labels=c('low','mid','high'))
```

## Derived variables and transformations

```r
#here we derived two variable for percenatge of earned hours to attempt hours for both the semester,
#FIRST_TERM_HRS_PERCENTAGE carry the same info as FIRST_TERM_EARNED_HRS and FIRST_TERM_ATTEMPT_HOURS we just reduced dimension
#and coercity of further data modelling
data<-mutate(data,FIRST_TERM_HRS_PERCENTAGE=data$FIRST_TERM_EARNED_HRS/data$FIRST_TERM_ATTEMPT_HRS,
       SECOND_TERM_HRS_PERCENTAGE=data$SECOND_TERM_EARNED_HRS/data$SECOND_TERM_ATTEMPT_HRS)
```

```r
data=as.data.frame(unclass(data))
data$RETURNED_2ND_YR=as.factor(data$RETURNED_2ND_YR)
```
